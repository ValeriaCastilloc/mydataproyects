## **Factor Analysis**


```python
df1 = df[['macha','machb','machc','machdrecode','mache','enga_t1','engb_t1','engc_t1','sila_t1','silb_t1','silc_t1','sild_t1','sile_t1']]
df1
```

## **Factor Analysis - Performing factor analysis (Initial Factor Analysis)**

```python
fa = FactorAnalyzer(rotation='varimax')
fa.fit(df1) # factor analysis on cleaned data
ev, v = fa.get_eigenvalues() # eigenvalues
ev
```

```python
sorted(ev) ##Syntax for sorting eigenvalue in ASC order
```

Here we recall that those eigenvalues higher than 1 are significant.
- We look for higher eigenvalues as a factor with a high eigenvalue means that it explains more of the variance in the data.

```python
# Scatter and line plot of eigenvalues to choose the number of significant factors
plt.scatter(range(1,df1.shape[1]+1),ev)
plt.plot(range(1,df1.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

As we can see here in the plot, Factors 1-3 seems to be the significant ones.

## **Factor Analysis - Performing factor analysis (Defined Factor Analysis)**

```python
pd.DataFrame(fa.loadings_, columns=['Factor1','Factor2','Factor3'],index=[df1.columns])
```

Interpretation notes:

1. Factor loadings represent the correlation coefficients between the observed variables and the latent factors and presented in each cell of this table.
2. High loadings, ones ≥ 0.5, indicate a strong relationship between the variable and the factor.
3. High loadings, ones < 0.5, indicate a weak relationship between the variable and the factor.


According to the results, we can highlight the following points:
- **Factor 1**:
Silence seems to be strongly related with Factor 1 as we see the following coefficients: sila_t1 (0.566), silb_t1 (0.871), silc_t1 (0.833), sild_t1 (0.780), sile_t1 (0.783).
- **Factor 2**:
Engagement seems to be strongly associated with Factor 2. The variables show strong coefficients as follows: enga_t1 (0.630), engb_t1 (0.818), engc_t1 (0.744).
- **Factor 3**:
Machiavellianism seems to be somehow strong-moderate associated with Factor 3. The variables coefficients Hence we will remove them to improve reliabilityare as follows: macha (0.499), machb (0.733), machc (0.537).

- Finally, machdrecode and mache variables have weak coefficients on all factors, which suggests that they might not strongly contribute.

```python
dft = df1.iloc[:, 0:5].astype(np.int64) # test
dft.head()
```

```python
# Syntax for calculating reliability
pg.cronbach_alpha(data=dft)
```

## **Create Constructs**

```python
df.head()
```

```python
## Syntax for converting data types to the float data type:

df = df.astype('float')
df.head()
```

```python
#Syntax for creating new column (variable) in df:

df['machiavellianism'] = df['macha'] + df['machb'] + df['machc']
df['machiav_avr'] = df['machiavellianism'] / 3
df.head() # to display the first few rows to verify
```

```python
df['engagement'] = df['enga_t1'] + df['engb_t1'] + df['engc_t1']
df['engage_avr'] = df['engagement'] / 3
df.head() # to display the first few rows to verify
```

```python
df['silence'] = df['sila_t1'] + df['silb_t1'] + df['silc_t1'] + df['sild_t1'] + df['sile_t1']
df['silence_avr'] = df['silence'] / 5
df.head() # to display the first few rows to verify
```

```python
df6 =df[['workexperience','age','nationality','gender','machiav_avr','engage_avr','silence_avr']]
df6
```

##**Preparation before running process analysis**

```python
# Removing Outliers to visualize distribution

def boxplot(column):
    sns.boxplot(data=df6,x=df6[f"{column}"])
    plt.title(f"Boxplot of {column}")
    plt.show()
```

```python
## Syntax for creating boxplots for multiple columns
boxplot('machiav_avr')
boxplot('engage_avr')
boxplot('silence_avr')
```

```python
#After visualizing the outliers using boxplots, now we can filter outliers out by selecting rows that meet specific criteria

df7 = df6[(df6['machiav_avr'] < 4.0) &
          (df6['engage_avr'] > 1.5) &
          (df6['silence_avr'] < 3.5)].copy()
```

```python
df7.describe()
```

## **Model 7**

```python
p1= Process(data=df7, model=7, x="machiav_avr", y="engage_avr", w="workexperience", m=["silence_avr"],controls=["age","nationality","gender"],center=True, boot=10000)
p1.summary()
```
